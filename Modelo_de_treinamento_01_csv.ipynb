```python
import pandas as pd

uri = "https://gist.githubusercontent.com/guilhermesilveira/2d2efa37d66b6c84a722ea627a897ced/raw/10968b997d885cbded1c92938c7a9912ba41c615/tracking.csv"
dados = pd.read_csv(uri) #passei os dados como variavel que recebe o csv
dados.head() # imprime as 3 primeiras colunas ....


```

```python
mapa = {
  "home" : "principal",
  "how_it_works" : "como_funciona",
  "contact": "contato", 
  "bought" : "comprou"
}
dados = dados = dados.rename(columns = mapa)
```

```python
x = dados [["principal","como_funciona","contato"]] #features dos meus dados...
x.head()
```

```python
y = dados [["comprou"]]
y.head()
```

```python
dados.shape
```

```python
treino_x = x [:75]
treino_y = y [:75]
teste_x = x[75:]
teste_y = y[75:]

print ("Treinamos com %d elementos" % (len)(treino_x), len(teste_x))
```

```python
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score

modelo = LinearSVC()
modelo.fit(treino_x, treino_y)
previsoes = modelo.predict(teste_x)

acuracia = accuracy_score(teste_y, previsoes) * 100
print("A acuracia foi %.2f%% " % acuracia)
```
